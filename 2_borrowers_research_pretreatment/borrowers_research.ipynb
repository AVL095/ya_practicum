С помощью функции проверим людей с отрицательным доходом

```python
# Создадим список категорий занятости
income_type_category = ['безработный','в декрете','госслужащий',
                       'компаньон','пенсионер','предприниматель',
                        'сотрудник','студент']

# создадим функцию number_incomes_negative,
# которая будет подсчитывать количество людей с отрицательным доходом, без учета NaN значений. 
# с двумя параметрами (df, income_type)
# в переменной income_list_negative сохраняются те строки таблицы df, для которых 
# значение в столбце 'days_employed' меньше нуля
# и одновременно значение в столбце 'income_type' равно параметру 'income_type'
# в переменной income_list_negative_count сохраняется число значений столбца 'days_employed',
# рассчитанное методом count() для income_list_negative
# функция возвращает значение income_list_negative_count
def number_incomes_negative(df, income_type):
    income_list_negative = df.loc[(df.loc[:,'days_employed'] < 0) & (df.loc[:,'income_type']==income_type)]
    income_list_negative_count = income_list_negative.loc[income_list_negative.loc[:,'days_employed']!='NaN']['days_employed'].count()
    return income_list_negative_count

# Выведем категории с количеством тех у кого отрицательный доход
for income_type in income_type_category:
    print ('{:<15}|'.format(income_type), '{:<2}'.format(number_incomes_negative(df, income_type))) 
```

Из общей картины выделяется категория "пенсионеры"  
получается, что их трудовой стаж положительный.  
Ранее было замечено, что положительные значения в столбце 'days_employed' имеют аномально большие значения.  
При этом известно, что средний трудовой стаж россиян при выходе на пенсию равен 34,5 года   (https://ria.ru/20131119/977986173.html) (или 34,5*365=12592,5 дней), округлим до 12593 дней.  

```python
# Создадим фун-цию аналогичную number_incomes_negative
# с условием 'days_employed <= 12593'
# для подсчета количества 'адекватных' значений трудового стажа

def number_incomes_normal(df, income_type):
    income_list_normal = df.loc[(df.loc[:,'days_employed'] <= 12593) & (df.loc[:,'income_type']==income_type)]
    income_list_normal_count = income_list_normal.loc[income_list_normal.loc[:,'days_employed']!='NaN']['days_employed'].count()
    return income_list_normal_count
```

Проведем лемматизацию значений столбца 'purpose' для выявления основых целей кредита

```python
from pymystem3 import Mystem
m = Mystem()

def lemmas(purpose):
    lemmas_list = ' '.join(m.lemmatize(purpose))
    return lemmas_list

df['lemma_purpose'] = df['purpose'].apply(lemmas)

from collections import Counter
Counter(df['lemma_purpose'])
```

Сгруппируем по 4-м основным категориям: "недвижимость", "автомобиль", "образование", "свадьба".

Модифицируем функцию lemmas дополнительными условиями в соответсвии с выявленными категориями

```python
def lemmas(purpose):
    lemmas_list = ' '.join(m.lemmatize(purpose))
    if 'недвижимость' in lemmas_list:
        return 'недвижимость'
    if 'жилье' in lemmas_list:
        return 'недвижимость'
    if 'автомобиль' in lemmas_list:
        return 'автомобиль'
    if 'свадьба' in lemmas_list:
        return 'свадьба'
    if 'образование' in lemmas_list:
        return 'образование'

    #return lemmas_list

df['lemma_purpose'] = df['purpose'].apply(lemmas)
# С помощью Counter подсчитаем уникальные значения целей кредита
from collections import Counter
# Добавим эти категории в столбец lemmas_purpose
Counter(df['lemma_purpose'])
```

Больше всего тех у кого нет детей, тех у кого детей > 2 мало. Поэтому
объединим людей по количеству детей в 3 группы: 
(нет детей), (менее 2-х детей) - включает тех у кого 1-2 ребенка, (более 2-х детей) - включает всех остальных у кого от 3 до 5 детей.

```python
def children_group(children):
    
        if children == 0:
                return 'нет детей'
        if children <= 2:
                return 'менее 2-х детей'
        return 'более 2-х детей'
# Применим функцию к столбцу children
df['children_group'] = df['children'].apply(children_group)
# Выведем первые 5 строк таблицы
df.head()
```

Рассчитаем среднее по столбцу значение дохода и создадим отдельные столбец mean_income_category с категориями дохода:
    'средний', 'выше среднего', 'ниже среднего'


```python
mean_income = df['total_income'].mean().astype('int')
def income_category(total_income):
    if total_income > mean_income:
        return 'больше среднего'
    if total_income < mean_income:
        return 'меньше среднего'
    else:
        return 'средний'

df['mean_income_category'] = df['total_income'].apply(income_category)    
```

Рассмотрим взаимосвязь целей кредита с возвратом в срок. 


```python
df_pivot_purpose = df.pivot_table(index=['lemma_purpose'], columns='debt', values='family_status_id', aggfunc='count')
df_pivot_purpose['percent'] = df_pivot_purpose[1]/(df_pivot_purpose[1]+df_pivot_purpose[0])
df_pivot_purpose['percent'] = df_pivot_purpose['percent'].map('{:.1%}'.format)
df_pivot_purpose.sort_values(by='percent', ascending=False)
```

Самый высокий процент невозврата наблюдается для автокредитов
не отстает просрочка по образовательным и кредитам на свадьбу
Самый низкий процент просрочек по жилищным кредитам, вероятно это связано с наибольшей "серьезностью" данной категории, и соответственно большей ответственностью заемщиков

Рассмотрим влияние количества детей и семейное положение на процент просрочки по кредиту

```python
df_pivot_children = df.pivot_table(index=['family_status','children_group'], columns='debt', values='family_status_id', aggfunc='count')
df_pivot_children = df_pivot_children.fillna(0)

df_pivot_children['percent'] = df_pivot_children[1]/(df_pivot_children[1]+df_pivot_children[0])
df_pivot_children['percent'] = df_pivot_children['percent'].map('{:.1%}'.format)
df_pivot_children.sort_values(by='family_status', ascending=False)
```
